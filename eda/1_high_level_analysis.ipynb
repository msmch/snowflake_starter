```python
import pandas as pd

import sys
sys.path.append('../')
from automation_scripts.builders import TableDDLBuilder, StageBuilder
from automation_scripts.file_operations import get_files, write_to_file
```

```python
# get file names
files = get_files('../data/')
```

```python
STORAGE_URL = 'azure://msmch.blob.core.windows.net/snowflake-starter'

for file in files:
    fpath = f'../data/{file}'
    df = pd.read_csv(fpath)

    print(file)
    display(df.head())
    print(df.info())

    core_name = file.split(".")[0]
    table_name = f'raw_{core_name}'
    tbl_builder = TableDDLBuilder(df)
    table_ddl = tbl_builder.build_statment(table_name)
    tbl_status = write_to_file(
        fpath = f'../warehouse/tables/staging/{table_name}.sql',
        content = table_ddl
    )
    print(tbl_status, '\n')

    stg_builder = StageBuilder(STORAGE_URL)
    stage_ddl = stg_builder.build_statement(core_name)
    stg_status = write_to_file(
        fpath = f'../warehouse/stages/stg_{core_name}.sql',
        content = stage_ddl
    )
    print(stg_status, '\n')
```
